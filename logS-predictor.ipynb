#prediction of logS


##load the dataset

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

#**data** **prep**

##separating data

```python
y = df['logS']
x = df.drop('logS', axis = 1) #axis=1 makes sure that it displays the data in form of columns
x
```

###split data as the training set and testing set


```python
from sklearn.model_selection import train_test_split as tspl
x_train, x_test, y_train, y_test = tspl(x , y, test_size = 0.2 , random_state=100)  #this means the size of test will be 20% of the total data and train size will be 80% od entire data.
```

#building the model

##traingin the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

##applying model to make a prediciton

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
y_lr_test_pred
```

#model performnace check
comparing predicted vals with actual data and seeing accuracy


```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)



```

```python
lr_res = pd.DataFrame(['Linearregression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
```

```python
lr_res.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_res.to_csv('lr_res.csv', index=False)
lr_res
```

#**main program**
taking input from user and output is teh predicted logS value

```python
print("enter values of MolLogP, MolWt, NumRotatableBonds, AromaticProportion to predict value of logS, separate using spaces and should be of type float")
mlp, mw, nrb, arp = input().split()

mlp = float(mlp)
mw = float(mw)
nrb = float(nrb)
arp = float(arp)

new_record = pd.DataFrame({
    'MolLogP': [mlp],
    'MolWt': [mw],
    'NumRotatableBonds': [nrb],
    'AromaticProportion': [arp]
})

# Step 3: Concatenate the new record with the existing DataFrame
df = pd.concat([df, new_record], ignore_index=True)

print("added to file succesfully")
new_rec = df.tail(1)[['MolLogP', 'MolWt', 'NumRotatableBonds', 'AromaticProportion']]
predval = lr.predict(new_rec)
print("\npredicted value is: logS= ", predval[0])
df.at[df.index[-1], 'logS'] = predval

```
